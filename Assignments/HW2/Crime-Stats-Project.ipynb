## Task 2 

* Create two `DataLoader` objects for training and testing based on the input and output variables. Pick a reasonable batch size and verify the shape of data by iterating over the one dataset and printing the shape of the batched data. 

```python

from torch.utils.data import TensorDataset

X = df[['X', 'Y', 'Area_Command', 'NIBRSOffenseCode', 'DayOfWeek', 'Time']].values
y = df['VictimCount'].values

X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.float32)

dataset = TensorDataset(X, y)

train_size = int(0.7 * len(dataset))
test_size = len(dataset) - train_size
train_dataset, test_dataset = torch.utils.data.random_split(dataset, [train_size, test_size])

batch_size = 60 

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

for batch in train_loader:
    inputs, targets = batch
    print("Batch shape - Inputs:", inputs.shape, "Targets:", targets.shape)
    break  

print("Number of training batches:", len(train_loader))
print("Number of testing batches:", len(test_loader))
```

## Task 3
In this task you will try to predict number of crime victims as a **real number**. Therefore the machine learning problem is a **regression** problem.

#3.1 Define the proper loss function for this task

The appropriate loss function is Mean Squared Error (MSE). MSE measures the average squared difference between the predicted values and the actual target values.

#3.2 what should the size of the predicted output be?

The size of the predicted output would be scalar, since we are predicting the victimCount.

#3.3 explain your choice of architecture, including how many layers you will be using?

Architecture used:

Input Layer: 6, same as the input size
Hidden Layers: 1 * 32 neurons
Output Layer: 1 neuron to pridict the victim count


```python

import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import mean_squared_error

class RegressionNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(RegressionNN, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.fc2 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = torch.relu(self.fc1(x))  # Using ReLU activation function
        x = self.fc2(x)
        return x

input_size = 6
hidden_size = 32
output_size = 1
learning_rate = 0.01
num_epochs = 10

model = RegressionNN(input_size, hidden_size, output_size)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

for epoch in range(num_epochs):
    model.train()
    
    for batch in train_loader:
        inputs, targets = batch
        
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, targets.view(-1, 1))
        loss.backward()
        optimizer.step()

    model.eval()
    total_mse = 0
    total_samples = 0
    
    with torch.no_grad():
        for batch in test_loader:
            inputs, targets = batch
            outputs = model(inputs)
            mse = mean_squared_error(outputs.numpy(), targets.numpy())
            batch_size = targets.size(0)
            total_mse += mse * batch_size
            total_samples += batch_size
    
    avg_mse = total_mse / total_samples
    
    print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {loss.item()}, Test MSE: {avg_mse}')

print('Training finished')


```

#4.1 How would you find the number of classes on the output data?

To determine the number of classes in the output data, We need to define the classification criteria. The number of classes will be the number of categories you define.

For example, Group number of crime victims into categories like "Low," "Medium," and "High" based on certain thresholds.

#4.2 How is the architecture different?

There will be a difference definitely since we are classifying the data. Difference would be with the output layer. it should be output layer with as many neurons as there are classes.

#4.3 How is the loss function different?

We are using MSE in the regression where we need to predict the continous numeric value, Where as in the classification Cross-Entrophy loss is used since the goal is to predict the class labels.

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np


class ClassificationNN(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super(ClassificationNN, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.fc2 = nn.Linear(hidden_size, num_classes)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x


X = df[['X', 'Y', 'Area_Command', 'NIBRSOffenseCode', 'DayOfWeek', 'Time']].values

df['VictimCount'] = pd.cut(df['VictimCount'], bins=[-1, 1, 5, np.inf], labels=['Low', 'Medium', 'High'])
y = df['VictimCount'].astype('category').cat.codes

X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.long)  # Long tensor for categorical values

dataset = TensorDataset(X, y)

train_size = int(0.8 * len(dataset))
test_size = len(dataset) - train_size
train_dataset, test_dataset = torch.utils.data.random_split(dataset, [train_size, test_size])

batch_size = 64
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

input_size = 6
hidden_size = 32
num_classes = len(df['VictimCount'].cat.categories)  # Number of unique categories

learning_rate = 0.001
num_epochs = 10

model = ClassificationNN(input_size, hidden_size, num_classes)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

for epoch in range(num_epochs):
    model.train()
    
    for batch in train_loader:
        inputs, targets = batch
        
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        loss.backward()
        optimizer.step()

    model.eval()
    correct = 0
    total_samples = 0
    
    with torch.no_grad():
        for batch in test_loader:
            inputs, targets = batch
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            total_samples += targets.size(0)
            correct += (predicted == targets).sum().item()
    
    accuracy = 100 * correct / total_samples
    
    print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {loss.item()}, Test Accuracy: {accuracy}%')

print('Training finished')
```

Worked on a regression problem in Task 3 with the goal of estimating the actual number of crime victims. As I moved on to Task 4, I had to anticipate the number of crime victims as a class label in a classification task.
```python
import matplotlib.pyplot as plt

model.eval()
misclassified_example = None

with torch.no_grad():
    for batch in test_loader:
        inputs, targets = batch
        outputs = model(inputs)
        probabilities = torch.softmax(outputs, dim=1)
        _, predicted = torch.max(outputs, 1)

        misclassified_mask = (predicted != targets)
        if misclassified_mask.any():
            misclassified_example = inputs[misclassified_mask][0]
            predicted_class = predicted[misclassified_mask][0].item()
            true_class = targets[misclassified_mask][0].item()
            break

if misclassified_example is not None:
    classes = list(df['VictimCount'].cat.categories)
    probabilities = probabilities[misclassified_mask][0].numpy()
    
    plt.bar(classes, probabilities)
    plt.xlabel('Class')
    plt.ylabel('Probability')
    plt.title(f'Misclassified Example\nPredicted: {classes[predicted_class]}, True: {classes[true_class]}')
    plt.show()
    
    print(f"Correct class: {classes[true_class]}")
```
